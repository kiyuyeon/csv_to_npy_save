```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

def convert_csv_to_npy(csv_dir, npy_dir):
    # npy_dir이 존재하지 않으면 생성
    if not os.path.exists(npy_dir):
        os.makedirs(npy_dir)

    # csv_dir에서 모든 CSV 파일을 찾아서 처리
    for csv_file in os.listdir(csv_dir):
        if csv_file.endswith('.csv'):
            # CSV 파일을 DataFrame으로 로드
            df = pd.read_csv(os.path.join(csv_dir, csv_file))
            
            data = df.values

            # DataFrame을 NPY 파일로 저장
            npy_file_name = csv_file.replace('.csv', '.npy')
            npy_file_path = os.path.join(npy_dir, npy_file_name)
            np.save(npy_file_path, data)
            
            print(f'Saved {npy_file_name} to {npy_dir}')

# 사용 예시
csv_dir = '/data/0ho/mobicare_pal_holter/destination_for_2'  # CSV 파일이 위치한 디렉토리
npy_dir = '/data/hw/esus_data/'  # NPY 파일을 저장할 디렉토리

convert_csv_to_npy(csv_dir, npy_dir)
```
